4- List all kidnapping crimes involving a child?

```python
%%sql 
--SELECT * FROM chicago_crime_data WHERE primary_type = 'KIDNAPPING'
SELECT * FROM chicago_crime_data 
    WHERE primary_type = 'KIDNAPPING' AND description LIKE '%CHILD%';
```

5- List the kind of crimes that were recorded at schools. (No repetitions)

```python
%%sql
SELECT DISTINCT(location), community_area, block FROM chicago_crime_data
    WHERE location_description LIKE '%SCHOOL%';
```

6- List the type of schools along with the average safety score for each type.

```python
%%sql
SELECT DISTINCT(elementary_or_high_school), AVG(safety_score)
    FROM school_chicago GROUP BY elementary_or_high_school;
```

7- List 5 community areas with highest % of households below poverty line

```python
%%sql
SELECT ca FROM chicago_socioeconomic_data
    ORDER BY percent_households_below_poverty DESC LIMIT 5;
```

8- Which community area is most crime prone? Display the coumminty area number only.

```python
res = %sql SELECT community_area, COUNT(community_area) AS cant_crim \
        FROM chicago_crime_data \
            GROUP BY community_area ORDER BY cant_crim DESC LIMIT 1;
print(res)
#communitry area most crimes es
ca_mc = res[0][0]
print(ca_mc)
```

9 - Use a sub-query to find the name of the community area with highest hardship index

```python
%%sql
SELECT community_area_name FROM chicago_socioeconomic_data
    WHERE hardship_index = (SELECT MAX(hardship_index) FROM chicago_socioeconomic_data)
```

10- Use a sub-query to determine the Community Area Name with most number of crimes?

```python
%%sql
SELECT community_area_name FROM chicago_socioeconomic_data
    WHERE ca = :ca_mc;
```
