```python
#Metodo with para conectarse al database y que se cierre la conexion automaticamente
from sqlalchemy import create_engine

DATABASE_URL_mysql = "mysql+mysqlconnector://root:toor@127.0.0.1:3306/exampleDB"
with create_engine(DATABASE_URL_mysql).connect() as mysql_conn:
    print(f"Connected to MySQL : {DATABASE_URL_mysql}")
    %sql $DATABASE_URL_mysql
    for i in range(15, 20):
        a = f"""
        INSERT INTO usersNuevo (user_first, user_last, user_email, user_uid, user_pwd)
        VALUES ('usuario_{i}','smith_{i}','smith_{i}@hotmail.com', 'smth_{i}', 'smth_{i*2}');
        """
        %sql $a
        
DATABASE_URL_postgres = "postgresql://root:toor@127.0.0.1:5432/company_db"
# PostgreSQL
with create_engine(DATABASE_URL_postgres).connect() as pg_conn:
    print(f"Connected to PostgreSQL: {DATABASE_URL_postgres}")
    %sql $DATABASE_URL_postgres
    for i in range(10, 15):
        b = f"""
        INSERT INTO jobs (job_id, job_title, min_salary, max_salary)
        VALUES ('{str(i)}','al_pedo_{i}',{7000*i}, {i*20000});
        """
        %sql $b
```
