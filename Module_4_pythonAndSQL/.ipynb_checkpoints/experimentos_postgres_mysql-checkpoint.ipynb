```python
from sqlalchemy import create_engine

DATABASE_URL_mysql = "mysql+mysqlconnector://root:toor@127.0.0.1:3306/exampleDB"

try:
    engine_mysql = create_engine(DATABASE_URL_mysql)
    connection_mysql = engine_mysql.connect()
    print(f"Connected successfully to {DATABASE_URL_mysql}!")
except Exception as e:
    print(f"Connection to {DATABASE_URL_mysql}failed:", e)
    
for i in range(15, 20):
    a = f"""
    INSERT INTO usersNuevo (user_first, user_last, user_email, user_uid, user_pwd)
    VALUES ('usuario_{i}','smith_{i}','smith_{i}@hotmail.com', 'smth_{i}', 'smth_{i*2}');
    """
    %sql $a

connection_mysql.close()
connection_mysql.invalidate()
engine_mysql.dispose()
```

```python
from sqlalchemy import create_engine

DATABASE_URL_mysql = "postgresql://root:toor@127.0.0.1:5432/company_db"
DATABASE_URL_postgres = "postgresql://root:toor@127.0.0.1:5432/company_db"

try:
    engine_postgres = create_engine(DATABASE_URL_postgres)
    connection_postgres = engine_postgres.connect()
    print(f"Connected successfully to {DATABASE_URL_postgres}!")
except Exception as e:
    print(f"Connection to {DATABASE_URL_postgres} failed:", e)

for i in range(4, 10):
    a = f"""
    INSERT INTO jobs (job_id, job_title, min_salary, max_salary)
    VALUES ('{str(i)}','al_pedo_{i}',{7000*i}, {i*20000});
    """
    %sql $a

connection_postgres.close()
engine_postgres.dispose()
```

```python
#Metodo with para conectarse al database y que se cierre la conexion automaticamente
from sqlalchemy import create_engine

DATABASE_URL_mysql = "mysql+mysqlconnector://root:toor@127.0.0.1:3306/exampleDB"
with create_engine(DATABASE_URL_mysql).connect() as mysql_conn:
    print(f"Connected to MySQL : {DATABASE_URL_mysql}")
    %sql $DATABASE_URL_mysql
    for i in range(15, 20):
        a = f"""
        INSERT INTO usersNuevo (user_first, user_last, user_email, user_uid, user_pwd)
        VALUES ('usuario_{i}','smith_{i}','smith_{i}@hotmail.com', 'smth_{i}', 'smth_{i*2}');
        """
        %sql $a
        
DATABASE_URL_postgres = "postgresql://root:toor@127.0.0.1:5432/company_db"
# PostgreSQL
with create_engine(DATABASE_URL_postgres).connect() as pg_conn:
    print(f"Connected to PostgreSQL: {DATABASE_URL_postgres}")
    %sql $DATABASE_URL_postgres
    for i in range(10, 15):
        b = f"""
        INSERT INTO jobs (job_id, job_title, min_salary, max_salary)
        VALUES ('{str(i)}','al_pedo_{i}',{7000*i}, {i*20000});
        """
        %sql $b
```

```python
#Metodo para conectarse y cerrar la conexion a mano
from sqlalchemy import create_engine

DATABASE_URL_mysql = "mysql+mysqlconnector://root:toor@127.0.0.1:3306/exampleDB"

try:
    engine_mysql = create_engine(DATABASE_URL_mysql)
    connection_mysql = engine_mysql.connect()
    %sql {str(engine_mysql.url)}
    print(f"Connected successfully to {str(engine_mysql.url)}!")
except Exception as e:
    print(f"Connection to {str(engine_mysql.url)} failed:", e)

for i in range(1, 10):
    a = f"""
        INSERT INTO usersNuevo (user_first, user_last, user_email, user_uid, user_pwd)
        VALUES ('usuario_{i}','smith_{i}','smith_{i}@hotmail.com', 'smth_{i}', 'smth_{i*2}');
        """
    %sql $a

connection_mysql.close()

DATABASE_URL_postgres = "postgresql://root:toor@127.0.0.1:5432/company_db"

try:
    engine_postgres = create_engine(DATABASE_URL_postgres)
    connection_postgres = engine_postgres.connect()
    %sql {str(engine_postgres.url)}
    print(f"Connected successfully to {str(engine_postgres.url)}!")
except Exception as e:
    print(f"Connection to {str(engine_postgres.url)} failed:", e)

for i in range(25, 30):
    a = f"""
    INSERT INTO jobs (job_id, job_title, min_salary, max_salary)
    VALUES ('{str(i)}','al_pedo_{i}',{7000*i}, {i*20000});
    """
    %sql $a

connection_postgres.close()
```
